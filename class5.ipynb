**Quiz**

```python
answer = input("When was the first known use of the word 'quiz'? ")
if answer == "1781":
    print("Correct!")
else:
    print(f"The answer is '1781', not {answer}")

answer = input("Which built-in function can get information from the user? ")
if answer == "input":
    print("Correct!")
else:
    print(f"The answer is 'input', not {answer}")
```

**Profit and loss**

```python

costprice =int(input("enter the cost price: "))
sellingprice =int(input("enter the selling price: "))

if(sellingprice>costprice):
  print("profit")
  pt=sellingprice-costprice
  print("$",pt)
else :
  print("No profit")

```

**odd or even**

```python
n=int(input("Enter the number:"))
if( n%2==0 ):
  print("The number is even")
else:
  print("The number is odd")

```

**Greater number**

```python
n1= int(input("Enter 1st number "))
n2= int(input("Enter 2nd number "))
if (n1>n2):
  print(n1, "is greater")
elif (n2>n1):
  print(n2," is greater")
else:
  print("Both are equal")

```
